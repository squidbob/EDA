```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("2015.csv")
df
```

```python
df.isna().any()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(data=df, x="Happiness Score", y="Health (Life Expectancy)", hue="Region")
plt.legend(bbox_to_anchor=(1.7, 0.5), loc="right", title="Region")
```

```python
westeur=df.loc[df["Region"]=="Western Europe"]
sahara=df.loc[df["Region"]=="Sub-Saharan Africa"]
westeur_sahara=pd.concat([westeur, sahara], axis=0)
westeur_sahara
```

```python
sns.histplot(data=westeur_sahara, x="Dystopia Residual", hue="Region", element="poly")
```

```python
sns.boxplot(data=df, x="Family", y="Region")
```
